```python
#Example using array "A" on how to change an array's axis
A = np.array([[1, 2, 5, 1]])
print(A)
print(A.transpose(1, 0))
```

```python
#Example using array "B" on sorting an array based on a column ("0" is column 1)
B = np.array([[1, 5],
                [2, 18],
                [9, 2],
                [1, 4]])
print(B)
print(B[B[:,0].argsort()])
```

```python
#Example using arrays "C" and "D" to connect two arrays together.
C = np.array([["Red", 7],
                ["Yellow", 10],
                ["Red", 0],
                ["Blue", 9]])
D = np.array([["Green", 4],
                ["Green", 3],
                ["Green", 6],
                ["Yellow", 2]])
print(C)
print(D)
print(np.concatenate((C,D)))
print(np.concatenate((D,C)))
```
